# Box Plot

```python
plt.figure(figsize=(15,10))
plt.subplot(2,2,1)
sns.boxplot(x='Classification',y='SL',data=data)
plt.subplot(2,2,2)
sns.boxplot(x='Classification',y='SW',data=data)
plt.subplot(2,2,3)
sns.boxplot(x='Classification',y='PL',data=data)
plt.subplot(2,2,4)
sns.boxplot(x='Classification',y='PW',data=data)
pass
```

# Strip Plot

```python
plt.figure(figsize=(15,10))
plt.subplot(2,2,1)
sns.stripplot(x=data['Classification'],y=data['SL'],data=data,hue='Classification')
plt.subplot(2,2,2)
sns.stripplot(x=data['Classification'],y=data['SW'],data=data,hue='Classification')
plt.subplot(2,2,3)
sns.stripplot(x=data['Classification'],y=data['PL'],data=data,hue='Classification')
plt.subplot(2,2,4)
sns.stripplot(x=data['Classification'],y=data['PW'],data=data,hue='Classification')
pass
```

# Heatmap

```python
sns.heatmap(data.corr(),annot=True)
plt.xticks(rotation=90)
```

# Violin Plot

```python
plt.figure(figsize=(15,10))
plt.subplot(2,2,1)
sns.violinplot(x='Classification',y='SL',data=data)
plt.subplot(2,2,2)
sns.violinplot(x='Classification',y='SW',data=data)
plt.subplot(2,2,3)
sns.violinplot(x='Classification',y='PL',data=data)
plt.subplot(2,2,4)
sns.violinplot(x='Classification',y='PW',data=data)
pass
```
